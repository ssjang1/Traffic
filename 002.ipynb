```python
import pandas as pd
data = pd.read_csv('EnglandWeather.csv')
data
```

```python
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'], utc=True, format='%Y-%m-%d %H:%M:%f%z', errors='coerce')
data
```

```python
data['year'] = data['Formatted Date'].dt.year
data['month'] = data['Formatted Date'].dt.month
data['day'] = data['Formatted Date'].dt.day
data['hour'] = data['Formatted Date'].dt.hour
data['minute'] = data['Formatted Date'].dt.minute
```

```python
data2 = data.dropna()
print(data.shape)
print(data2.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 필요한 라이브러리를 불러옵니다.
# 먼저 수치형 데이터만 선택합니다.
features = ['Temperature (C)', 'Wind Speed (km/h)', 'Pressure (millibars)', 'Humidity', 'year', 'month', 'day', 'hour', 'minute']
x = data2.loc[:, features].values

# 데이터를 표준화합니다.
x = StandardScaler().fit_transform(x)

# PCA를 적용합니다.
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
```

```python
from sklearn.manifold import TSNE

# t-SNE 모델을 생성하고 적용합니다.
tsne = TSNE(n_components=3, random_state=0)
x_tsne = tsne.fit_transform(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# t-SNE로 차원 축소한 데이터 x_tsne 준비
# 여기서 x_tsne는 (샘플 수, 3) 모양을 가집니다.
# y는 샘플의 레이블

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')  # 3D 그래프를 위한 축 추가

# 3차원 산점도 그리기
scatter = ax.scatter(x_tsne[:, 0], x_tsne[:, 1], x_tsne[:, 2], c=y, cmap='viridis')

# 축 레이블 설정
ax.set_xlabel('Axis 1')
ax.set_ylabel('Axis 2')
ax.set_zlabel('Axis 3')  

# 컬러바 추가
plt.colorbar(scatter)

# 제목 설정
plt.title('t-SNE Results')

plt.savefig('t-SNE_results.jpg', bbox_inches='tight')
plt.show()
```

```python
import umap
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# UMAP 모델을 생성하고 3차원으로 적용합니다.
reducer = umap.UMAP(n_components=3, random_state=42)
x_umap_3d = reducer.fit_transform(x)

# 3D 시각화
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3차원 산점도 그리기
scatter = ax.scatter(x_umap_3d[:, 0], x_umap_3d[:, 1], x_umap_3d[:, 2], c=y, cmap='viridis')

# 컬러바 추가
plt.colorbar(scatter)

# 제목 및 축 레이블 설정
ax.set_title('3D UMAP Results')
ax.set_xlabel('Axis 1')
ax.set_ylabel('Axis 2')
ax.set_zlabel('Axis 3')

plt.savefig('umap_results.jpg', bbox_inches='tight')
plt.show()
```
