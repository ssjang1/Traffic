```python
data['SUM_YRMTHDAT'] = data['SUM_YRMTHDAT'].astype('str')
data['SUM_YRMTHDAT'] = pd.to_datetime(data['SUM_YRMTHDAT'], format="%Y%m%d")
```

```python
data
```

```python
data['year'] = data['SUM_YRMTHDAT'].dt.year
data['month'] = data['SUM_YRMTHDAT'].dt.month
data['day'] = data['SUM_YRMTHDAT'].dt.day
data['hour'] = data['SUM_HR']
```

```python
fig1 = plt.figure(figsize=(12,5), dpi=100) 

axe1 = fig1.add_axes([0, 0, 1, 1]) 

axe1.scatter(data['hour'], data['TRFFCVLM'], color='blue', marker='o', s=10) 

axe1.set_xlabel('hour', fontsize=12) 

axe1.set_ylabel('Trffcvlm', fontsize=12) 

axe1.set_title('Hour vs. Trffcvlm', fontsize=20) 

axe1.grid(False) 

axe1.set_xlim(data['hour'].min()-1, data['hour'].max()+1) 

axe1.set_xticks(np.arange(data['hour'].min(), data['hour'].max()+1, 1)) 

plt.show()
```

```python
fig2 = plt.figure(figsize=(12,5), dpi=100)
axe2 = fig2.add_axes([0, 0, 1, 1])
axe2.set_xticks(np.arange(data['TRFFCVLM'].min(), data['TRFFCVLM'].max()+1, 100))

axe2.scatter(data['TRFFCVLM'], data['SPD_AVG'], color='blue',marker='o', s=10)

axe2.set_xlabel('TRFFCVLM', fontsize=12)
axe2.set_ylabel('SPD_AVG', fontsize=12)
axe2.set_title('TRFFCVLM vs. SPD_AVG', fontsize=20)
axe2.grid(False)
plt.xticks(rotation=90)
axe2.set_xlim(data['TRFFCVLM'].min()-1, data['TRFFCVLM'].max()+1)

plt.show()
```

```python
plt.figure(figsize=(15,5), dpi=100)
plt.scatter(data["TRFFCVLM"], data["SPD_AVG"], c = data['OCCPNCY'], cmap='viridis_r', marker='o', s=10, alpha=0.5) 

plt.xticks(np.arange(data['TRFFCVLM'].min(), data['TRFFCVLM'].max()+1, 100), rotation=90)

plt.title("TRFFCVLM vs. SPD_AVG", fontsize = 20)
plt.xlabel("TRFFCVLM", fontsize = 20)
plt.ylabel("SPD_AVG", fontsize = 20)
plt.grid(False)
plt.colorbar().set_label('OCCPNCY', fontsize = 20)

plt.savefig('교통량_속도_점유율.jpg',bbox_inches='tight')
plt.show()
```

```python
test_data = data.loc[:,['TRFFCVLM','OCCPNCY','SPD_AVG','hour']]
test_data
```

```python
hourly_mean = test_data.groupby(by='hour').mean()
hourly_mean
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(20, 10), dpi=100)

# ax1을 사용하여 첫 번째 데이터 플롯
ax1.plot(hourly_mean.index, hourly_mean['SPD_AVG'], lw=2, color="blue", marker='o', label="SPD_AVG")

ax1.set_xlabel('hour', fontsize=18)
ax1.set_ylabel("SPD_AVG", fontsize=18, color="blue")
ax1.legend(loc='upper left', fontsize=13, bbox_to_anchor=(0,1))
ax1.set_title("Average Values")


# ax1의 오른쪽 y축을 사용하여 두 번째 데이터 플롯
ax2 = ax1.twinx()
ax2.plot(hourly_mean.index, hourly_mean['TRFFCVLM'], lw=2, color="red", marker='v', label="TRFFCVLM")
ax2.set_ylabel("TRFFCVLM", fontsize=18, color="red")
ax2.legend(loc='upper left', fontsize=13, bbox_to_anchor=(0,0.95))

# ax1에서 새로운 ax3 생성 후, 위치 조정
ax3 = ax1.twinx()
# ax3의 위치를 오른쪽에서 조금 더 밀어내기
ax3.spines['right'].set_position(('outward', 60))  # 60은 축의 오프셋 거리입니다; 조정이 필요할 수 있습니다.
ax3.plot(hourly_mean.index, hourly_mean['OCCPNCY'], lw=2, color="green", marker='*', label="OCCPNCY")
ax3.set_ylabel("OCCPNCY", fontsize=18, color="green")
ax3.legend(loc='upper left', fontsize=13, bbox_to_anchor=(0,0.9))

plt.tight_layout()
plt.savefig('교통량_속도_점유율_평균값.jpg', bbox_inches='tight')
plt.show()
```
